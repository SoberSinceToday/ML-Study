**Алгоритмы снижения размерности**

**Импорты**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
import sklearn.datasets as datasets
from sklearn.preprocessing import StandardScaler

from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
import umap.umap_ as umap
```

**Считывание данных**

```python
digits = datasets.load_digits()
X = digits.data  # Признаки
y = digits.target  # Целевые метки

def visualise(x1, x2):
    plt.figure(figsize=(8, 6))
    plt.scatter(x1, x2, c=y, cmap='jet', edgecolor='k', s=40)
    plt.colorbar()
    plt.title("Визуализация")
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.show()
```

**PCA**

Линейный алгоритм

Уменьшает размерность данных с максимально возможным сохранением дисперсии

Дисперсия - разброс данных относительно среднего значения

• Сохраняет глобальную структуру

```python
from sklearn.preprocessing import StandardScaler

X_scaled = StandardScaler().fit_transform(X)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

visualise(X_pca[:, 0], X_pca[:, 1])
```

**t-SNE**

Нелинейный алгоритм

• Только 2-x и 3-x мерные данные

Преобразует сходства между точками данных в совместные вероятности и минимизирует расхождение между ними в различных пространствах, что позволяет выявлять кластеры внутри данных.

• Сохраняет локальное расстояние между точками

```python
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

visualise(X_tsne[:, 0], X_tsne[:, 1])
```

**UMAP**

Нелинейный алгоритм

Сохраняет как глобальную, так и локальную структуры

Значительно быстрее t-SNE

```python
umap_model = umap.UMAP(n_components=2, random_state=42)
X_umap = umap_model.fit_transform(X)

visualise(X_umap[:, 0], X_umap[:, 1])
```
